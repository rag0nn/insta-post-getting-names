```python
file = open("comments.txt","r",encoding='utf-8')
lines = file.readlines()
```

```python
row = []
rows = []

row_counter = 0
for line in lines:
    if line == " \n":
        rows.append(row)
        row_counter+=1
        print(row_counter," ",row)
        row = []
    else:
        row.append(line)
        
        
```

```python
print(len(rows))
```

```python
row_counter = 0

semacounter = 0

names = []
for row in rows:
    row_counter += 1
    print(row_counter," ",row[-1][:-1])
    names.append(row[-1][:-1])
    if "semaaaaks" in row[-1][:-1]: 
        semacounter +=1
        print(semacounter)
```

```python
len(names)
```

```python
temp_c = 0
for name in names:
    if "semaaaaks" in name:
        temp_c+=1
print(temp_c)
```

```python
newf = open("names.txt","w",encoding='utf-8')
for name in names:
    newf.writelines(name+"\n")
```
